```python
from langchain import PromptTemplate

# 定义模板
template = """作为一个广告专家，请为以下产品提供一个广告文案: {product}"""

# 创建提示模板
prompt = PromptTemplate(
    input_variables=["product"],
    template=template
)

# 格式化提示
formatted_prompt = prompt.format(product="华为Mate60 Pro手机")

# 打印格式化后的提示
print(formatted_prompt)
```

```python
from langchain.prompts import FewShotPromptTemplate, PromptTemplate
from langchain.schema import HumanMessage
from langchain_community.chat_models import ChatTongyi
import os
from dotenv import load_dotenv

# 加载环境变量
load_dotenv()

# 初始化聊天模型
chat = ChatTongyi(
    model='qwen-plus',
    top_p=0.9,
    temperature=0.9,
    api_key=os.getenv("DASHSCOPE_API_KEY")
)

# 创建示例
examples = [
    {"input": "您喜欢的手机是什么？", "output": "我喜欢的手机是iPhone。"},
    {"input": "您的喜欢手机的价格段位在多少？", "output": "我的预算在3000到5000元之间。"},
    {"input": "您对手机改进任何建议吗？", "output": "我希望电池续航能够更长一些。"},
]

# 定义格式化模板
example_formatter_template = """问题: {input} 答案: {output}\n"""
example_prompt = PromptTemplate(
    input_variables=["input", "output"],
    template=example_formatter_template,
)

# 定义后缀
suffix = "问题: {input_var}\n答案:"

# 创建 FewShotPromptTemplate 实例
few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    prefix="请回答以下问题：",
    suffix=suffix,
    input_variables=["input_var"],
    example_separator="\n\n",
)

# 定义待询问问题
input_question = "您最喜欢的手机颜色是什么？"
formatted_prompt = few_shot_prompt.format(input_var=input_question)

# 将格式化后的提示包装为 HumanMessage，并传递给 chat()
response = chat([HumanMessage(content=formatted_prompt)])

# 打印模型的回复
print(response.content)
```
